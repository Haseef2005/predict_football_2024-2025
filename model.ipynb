```python
# ML work with numeric data only
# 1 convert date
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
matches_df = pd.get_dummies(df, columns=['venue'], dtype=int, drop_first=True)
matches_df.head()
```

We will use category codes like 0 1 2 3 4 and so on for opponent code

We will not use getdummies for sure because it so many team

```python
matches_df['opp_code'] = matches_df['opponent'].astype('category').cat.codes
matches_df.head()
```

```python
matches_df['opp_code'].value_counts()
```

```python
team_code = matches_df[['opponent', 'opp_code']].drop_duplicates()
team_code.to_csv('team_code.csv')
```

```python
matches_df['hour'] = matches_df['time'].str.replace(':.+', '', regex=True).astype('int')
```

```python
matches_df.head()
```

```python
matches_df['day_code'] = matches_df['date'].dt.day_of_week
matches_df.head(5)
```

```python
matches_df['target'] = matches_df['result']
mapping_result = {
    'W' : 1,
    'D' : 0,
    'L' : 0,
}
matches_df['target'] = matches_df['target'].map(mapping_result)
matches_df.head()
```

```python
# we use classifier model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

predictors = ['venue_Home', 'opp_code', 'hour', 'day_code']
target = ['target']
X = matches_df[predictors]
y = matches_df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=141)

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=dict(
        n_estimators=[50, 100, 200, 300, 400, 500], 
        min_samples_split=[2, 5, 10, 15, 20, 25, 30, 40],  
        min_samples_leaf=[1,10,20,30],      
        random_state=[1],
    ),                                      
        scoring='f1_weighted',             
        cv=5,                                
        n_jobs=-1                            
)
grid_search.fit(X_train, y_train)

model = grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
predictions = model.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions,digits=4))
```

```python
print(confusion_matrix(y_test,predictions))
```

Improve precision with rolling averages

```python
grouped_matches = matches_df.groupby('team')
group = grouped_matches.get_group('Liverpool')
group.head()
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    # closed = 'left' is take the current week out
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt', 'xg', 'xga']
new_cols = [f'{c}_rollinig' for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches_df.groupby('team').apply(lambda x : rolling_averages(x,cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

predictors = ['venue_Home', 'opp_code', 'hour', 'day_code']
new_predtictors = predictors + new_cols
target = ['target']
X = matches_rolling[new_predtictors]
y = matches_rolling[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=1)

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=dict(
        n_estimators=[50, 100, 200], 
        min_samples_split=[2, 5],  
        min_samples_leaf=[1, 10],      
        random_state=[1],
    ),                                      
        scoring='f1_weighted',             
        cv=5,                                
        n_jobs=-1                            
)
grid_search.fit(X_train, y_train)

RFmodel = grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
predictions = RFmodel.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions,digits=4))
```

XGBoost

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

predictors = ['venue_Home', 'opp_code', 'hour', 'day_code']
new_predtictors = predictors + new_cols
target = ['target']
X = matches_rolling[new_predtictors]
y = matches_rolling[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=1)

param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 200],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'scale_pos_weight': [1, 2]
}
grid_search = GridSearchCV(
    estimator=XGBClassifier(random_state=1),
    param_grid=param_grid,                                      
    scoring='f1_weighted',             
    cv=5,                                
    n_jobs=-1                            
)
grid_search.fit(X_train, y_train)

XGBoostmodel = grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
predictions = XGBoostmodel.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions,digits=4))
```

```python
print(confusion_matrix(y_test,predictions))
```
